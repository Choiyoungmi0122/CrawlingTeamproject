```python
import pandas as pd

# CSV 파일 읽기
file = pd.read_csv('/Users/kim/Desktop/youngmi/0603/강서구1.csv')

# 요일 컬럼 이름 생성
days = ["월", "화", "수", "목", "금", "토", "일"]

# 필요한 컬럼 선택
columns_to_keep = ["name", "type", "phone", "address", "월", "화", "수", "목", "금", "토", "일"]
file = file[columns_to_keep]

# 정기휴무인 요일의 "(정기휴무)" 제거
file['holiday'] = ''
for day in days:
    file[day] = file[day].str.replace(r'\(정기휴무\)', '')
    file.loc[file[day] == '정기휴무', 'holiday'] = day


# CSV 파일로 저장
file.to_csv('/Users/kim/Desktop/youngmi/0603/강서구3.csv', index=False, encoding='UTF-8')
```

```python
import pandas as pd

# CSV 파일 읽기
file = pd.read_csv('/Users/kim/Desktop/youngmi/0603/강서구1.csv')

# 요일 컬럼 이름 생성
days = ["월", "화", "수", "목", "금", "토", "일"]

# 필요한 컬럼 선택
columns_to_keep = ["name", "type", "phone", "address", "월", "화", "수", "목", "금", "토", "일"]
file = file[columns_to_keep]

# "Mon" 컬럼을 "work"와 "break"으로 분할
file[['work', 'break']] = file['월'].str.split('\n', expand=True, n=1)
print(file)

# CSV 파일로 저장
file.to_csv('/Users/kim/Desktop/youngmi/0603/강서구workbreak.csv', index=False, encoding='UTF-8')

# "break" 컬럼 정리 및 "deadtime" 컬럼 생성
file["break"] = file["break"].str.replace("휴게시간", "")
file["deadtime"] = file["break"].str.extract(r"\n(.+?) 접수마감")
file["deadtime"] = file["deadtime"].str.replace("접수마감", "")
# "break" 컬럼에서 \n 이후의 값을 삭제
file["break"] = file["break"].str.split("\n").str[0]


file[["name", "type", "phone", "work", "break"]].head(10)
```
