```python
import csv
import json
import pandas as pd
import numpy as np

# JSON 파일 열기
with open('/Users/kim/Desktop/youngmi/json모아둠/남구.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)

# 변환된 데이터를 담을 리스트 생성
transformed_data = []

# 각 객체에 대해 변환 수행
for item in data['병원정보']:
    working_hours = item['영업시간']
    working_hours_dict = {}
    for working_hour in working_hours:
        day, hours = working_hour.split('\n', 1)
        working_hours_dict[day] = hours.strip()
    
    transformed_item = {
        '병원이름': item['이름'],
        '병원주소': item['병원주소'],
        '병원종류': item['병원종류'],
        **working_hours_dict,
        '전화번호': item['전화번호']
    }
    transformed_data.append(transformed_item)

# CSV 파일로 저장
all_keys = set().union(*(item.keys() for item in transformed_data))
fieldnames = list(all_keys)

with open('/Users/kim/Desktop/youngmi/0605/남구.csv', 'w', encoding='utf-8-sig', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(transformed_data)


# CSV 파일 읽기
file = pd.read_csv('/Users/kim/Desktop/youngmi/0605/남구.csv')


# 요일 컬럼 이름 생성
days = ["월", "화", "수", "목", "금", "토", "일"]

# 필요한 컬럼 선택
columns_to_keep = ["병원이름", "병원종류", "전화번호", "병원주소", "월", "화", "수", "목", "금", "토", "일"]
file = file[columns_to_keep]


# 'nan' 값을 빈 문자열로 대체
file = file.replace(np.nan, '', regex=True)

# '영업시간'과 '쉬는시간' 열 생성
file['영업시간'] = file[['월', '화', '수', '목', '금', '토', '일']].apply(lambda x: '\n'.join([str(i).split('\n')[0] if str(i) != '' else '' for i in x]), axis=1)
file['쉬는시간'] = file[['월', '화', '수', '목', '금', '토', '일']].apply(lambda x: '\n'.join([str(i).split('\n')[1] if len(str(i).split('\n')) > 1 and str(i) != '' else '' for i in x]), axis=1)



# "break" 컬럼 정리 및 "deadtime" 컬럼 생성
file["쉬는시간"] = file["쉬는시간"].str.replace("휴게시간", "")
file["접수마감"] = file["쉬는시간"].str.extract(r"\n(.+?) 접수마감")
file["접수마감"] = file["접수마감"].str.replace("접수마감", "")
# "쉬는시간" 컬럼에서 \n 이후의 값을 삭제
file["쉬는시간"] = file["쉬는시간"].str.split("\n").str[0]

# 정규식을 사용하여 "(매주 ...요일)" 형식의 문자열을 추출하여 정기휴무 컬럼에 할당
file["정기휴무"] = file["영업시간"].str.extract(r"\(매주(.+?)\)")

# 각 행의 정기휴무가 여러 개인 경우 줄바꿈 문자를 삽입하여 값 할당
file["정기휴무"] = file["정기휴무"].str.replace(", ", "\n")

# CSV 파일로 저장
file.to_csv('/Users/kim/Desktop/youngmi/0603/남구1.csv', index=False, encoding='utf8')

file[["영업시간"]].count
```
