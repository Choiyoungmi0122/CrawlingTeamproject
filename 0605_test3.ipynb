```python
import pandas as pd

# CSV 파일 읽기
file = pd.read_csv('/Users/kim/Desktop/youngmi/0603/강서구1.csv')

# 요일 컬럼 이름 생성
days = ["월", "화", "수", "목", "금", "토", "일"]

# 필요한 컬럼 선택
columns_to_keep = ["name", "type", "phone", "address"] + days
file = file[columns_to_keep]

# 정기휴무인 요일의 "(정기휴무)" 제거
file['holiday'] = ''
for day in days:
    file[day] = file[day].str.replace(r'\(정기휴무\)', '')
    file.loc[file[day] == '정기휴무', 'holiday'] = day
    file.loc[file[day].isnull(), 'holiday'] = ''

# work와 break 컬럼 생성
for day in days:
    file[f'{day}_work'] = file[day].str.split('\n').str[0]
    file[f'{day}_break'] = file[day].str.split('\n').str[1:]

file = file.drop(columns=days)

# CSV 파일로 저장
file.to_csv('/Users/kim/Desktop/youngmi/0603/강서구요일별일시간.csv', index=False, encoding='utf-8-sig')

file[["name", "type", "phone", "work", "break"]].head(10)
```

```python
import pandas as pd

# CSV 파일 읽기
file = pd.read_csv('/Users/kim/Desktop/youngmi/0603/강서구1.csv')

# 요일 컬럼 이름 생성
days = ["월", "화", "수", "목", "금", "토", "일"]

# 필요한 컬럼 선택
columns_to_keep = ["name", "type", "phone", "address"] + days
file = file[columns_to_keep]

# 정기휴무인 요일의 "(정기휴무)" 제거
file['holiday'] = ''
for day in days:
    file[day] = file[day].str.replace(r'\(정기휴무\)', '')
    file.loc[file[day] == '정기휴무', 'holiday'] = day
    file.loc[file[day].isnull(), 'holiday'] = ''

# work와 break 컬럼 생성
for day in days:
    file[f'{day}_work'] = file[day].str.split('\n').str[0]
    file[f'{day}_break'] = file[day].str.split('\n').str[1:].str.join(' ')

# CSV 파일로 저장
file.to_csv('/Users/kim/Desktop/youngmi/0603/강서구요일별일시간.csv', index=False, encoding='utf-8-sig')

file[["name", "type", "phone", "월_work", "월_break"]].head(10)
```
